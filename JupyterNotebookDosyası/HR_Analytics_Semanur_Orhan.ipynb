### Temel İstatistiklerin ve Dağılımların Analizi

**Veri Setindeki Tekrarlanan Satırların Tespiti ve Düzeltilmesi**

```python
duplicated_rows = df[df.duplicated(subset='EmpID', keep=False)]
print(f"Tekrar eden satır sayısı: {duplicated_rows.shape[0]}")
duplicated_rows.sort_values('EmpID')
```

```python
df_combined = df.groupby('EmpID').agg(lambda x: x.mean() if x.dtype == 'float64' else x.mode().iloc[0])
print(f"Birleştirilmiş veri setindeki satır sayısı: {df_combined.shape[0]}")
df_combined.head()
```

```python
# Son halini görüntüleyelim
df_combined.info()
df_combined.head()
```

**Veri Setindeki Eksik ve Aykırı Değerlerin Tespiti**

* **Eksik değer (NaN)** içeren satırları bulmak

```python
missing_data_rows = df[df.isnull().any(axis=1)]
print("Eksik Değer İçeren Satırlar:\n", missing_data_rows)
```

* NaN değerleri temizlemek

```python
df_cleaned = df.dropna()
```

* Kalan NaN değer kontrolü

```python
remaining_missing_values = df_cleaned.isnull().sum().sum()
print("\nSilme işleminden sonra kalan eksik (NaN) değer sayısı:", remaining_missing_values)
```

* **Aykırı Değer Kontrolü**: IQR (Interquartile Range) aykırı değer yakalamak için en sık kullanılan yöntemdir. Aşağıdaki kod bloğunda bu yöntem ile aykırı değerler için işlemler bulunmaktadır.
IQR yöntemi ile ilgili daha fazla bilgiye [buradan](https://online.stat.psu.edu/stat200/lesson/3/3.2) ulaşabilirsiniz.

```python
# Sadece sayısal değer sütunları seçilir
numeric_columns = df_cleaned.select_dtypes(include=['float64', 'int64'])

# Aykırı değerleri kontrol etmek için IQR yöntemini kullanılır
Q1 = numeric_columns.quantile(0.25)
Q3 = numeric_columns.quantile(0.75)
IQR = Q3 - Q1

# Aykırı değerleri belirlenir
is_outlier = (numeric_columns < (Q1 - 1.5 * IQR)) | (numeric_columns > (Q3 + 1.5 * IQR))

# Aykırı olmayan değerlerin bulunduğu satırlar seçilir
df_no_outliers = df_cleaned[~is_outlier.any(axis=1)]

# Temizlenmiş veri setinin boyutunu kontrol edilir
print("Yeni veri setinin boyutu:", df_no_outliers.shape)
```

---
**Temel İstatistiklerinin Elde Edilmesi**

* Mean hesaplanması

```python
mean_values = df_cleaned.mean(numeric_only=True)
print("Ortalama Değerler:\n",mean_values)
```

---
* Median hesaplanması

```python
median_values = df_cleaned.median(numeric_only=True)
print("\nMedyan Değerler:\n", median_values)
```

---
* Mod hesaplanması

```python
mode_values = df_cleaned.mode(numeric_only=True).iloc[0]
print("\nMod Değerler:\n", mode_values)
```

---
* Standart Sapma hesaplanması

```python
std_values = df_cleaned.std(numeric_only=True)
print("\nStandart Sapmalar:\n", std_values)
```
